```python
import pandas as pd
from sqlalchemy import create_engine

# swap in your postgres password; db name is whatever you called it
engine = create_engine("postgresql+psycopg2://postgres:idk@123R@localhost:5432/supplychain")

query = """
SELECT p.product_name,
       date_trunc('month', o.order_date)::date AS order_month,
       SUM(i.quantity) AS demand
FROM fact_order_items i
JOIN dim_product p ON p.product_id = i.product_id
JOIN fact_orders o ON o.order_id  = i.order_id
GROUP BY p.product_name, date_trunc('month', o.order_date)
ORDER BY p.product_name, order_month;
"""

df = pd.read_sql(query, engine, parse_dates=["order_month"])
print(df.shape)
df.head()
```
